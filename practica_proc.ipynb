```
# File location and type
file_location = "dbfs:/FileStore/big-data-proc/world_happiness_report.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Renombrar las columnas
df = df.withColumnRenamed("Country name", "country")
df = df.withColumnRenamed("Life Ladder", "ladder")
df = df.withColumnRenamed("Log GDP per capita", "gdp")
df = df.withColumnRenamed("Healthy life expectancy at birth", "esperanza_vida")


# Mostrar el DataFrame con las columnas renombradas
display(df)

# Crear vista temporal
temp_table_name2 = "tabla_anios"
df.createOrReplaceTempView(temp_table_name2)
```

```
%sql

SELECT country
FROM tabla_2021
ORDER BY ladder DESC
LIMIT 1;
```

```
%sql

SELECT country
FROM tabla_2021
WHERE region IN ('Western Europe', 'Central and Eastern Europe')
ORDER BY ladder DESC
LIMIT 1;

SELECT country
FROM tabla_2021
WHERE region IN ('Latin America and Caribbean', 'North America and ANZ')
ORDER BY ladder DESC
LIMIT 1;

SELECT country
FROM tabla_2021
WHERE region IN ('Middle East and North Africa', 'Sub-Saharan Africa')
ORDER BY ladder DESC
LIMIT 1;

SELECT country
FROM tabla_2021
WHERE region IN ('South Asia', 'East Asia','Southeast Asia','Commonwealth of Independent States')
ORDER BY ladder DESC
LIMIT 1;
```

```
%sql 

SELECT region, country
FROM (
    SELECT region, country,
           ROW_NUMBER() OVER (PARTITION BY region ORDER BY ladder DESC) AS ranking
    FROM tabla_2021
) AS subconsulta
WHERE ranking = 1;
```

```
%sql

SELECT country, COUNT(*) AS veces_primero
FROM (
    SELECT country, ladder,
           ROW_NUMBER() OVER (PARTITION BY year ORDER BY ladder DESC) AS ranking
    FROM tabla_anios
) AS subconsulta
WHERE ranking = 1
GROUP BY country
ORDER BY veces_primero DESC
LIMIT 1;
```

```
%sql

SELECT country,gdp
FROM tabla_anios
WHERE year = '2020'
ORDER BY gdp DESC
LIMIT 1;
```
